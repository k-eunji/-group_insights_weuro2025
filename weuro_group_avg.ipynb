```python
import pandas as pd

# 데이터 불러오기 (파일명은 round1_3.csv로 가정)
df = pd.read_csv('round1_3.csv')

# HomeTeam과 AwayTeam으로 나눠서 각 팀별 값 정리
# Home팀과 Away팀의 컬럼 이름 차이 제거를 위해 컬럼명을 통일해줌
home_cols = ['HomeGoals', 'HomeShots', 'HomeShotsOnTarget', 'HomexG', 'HomeBigChances', 'HomePossession(%)', 'HomePasses(%)',
             'HomeFouls', 'HomeTackles(%)', 'HomeClearances', 'HomeInterceptions', 'HomeErrorsLeadingtoShot', 'HomeErrorsLeadingtoGoal',
             'HomeSaves', 'HxGOTfaced', 'HGoalsprevented']

away_cols = ['AwayGoals', 'AwayShots', 'AwayShotsOnTarget', 'AwayxG', 'AwayBigChances', 'AwayPossession(%)', 'AwayPasses(%)',
             'AwayFouls', 'AwayTackles(%)', 'AwayClearances', 'AwayInterceptions', 'AwayErrorsLeadingtoShot', 'AwayErrorsLeadingtoGoal',
             'AwaySaves', 'AxGOTfaced', 'AGoalsprevented']

# Home 데이터 프레임
home_df = df[['HomeTeam'] + home_cols].copy()
home_df.columns = ['Team'] + [col.replace('Home', '') for col in home_cols]

# Away 데이터 프레임
away_df = df[['AwayTeam'] + away_cols].copy()
away_df.columns = ['Team'] + [col.replace('Away', '') for col in away_cols]

# 두 데이터 합치기
team_stats = pd.concat([home_df, away_df])

# 팀별 각 컬럼 평균 계산
team_avg = team_stats.groupby('Team').mean()

# 결과 출력
pd.options.display.float_format = '{:.2f}'.format
print(team_avg)

team_avg.to_csv("wteam_avg.csv")
```
